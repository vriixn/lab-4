detect outliers

```python

df.loc[len(df)] = [201, 'Male', 120, 300, 10]
df.loc[len(df)] = [202, 'Female', 10, 5, 99]

Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers_age = df[(df['Age'] < lower) | (df['Age'] > upper)]
outliers_age
```

```python
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers_age = df[(df['Age'] < lower) | (df['Age'] > upper)]
outliers_age
```


An outlier was detected in the Age column with a value of 120. This value was artificially added to demonstrate the outlier detection process using the IQR method. It represents an unusually high age compared to the rest of the dataset and is considered an extreme value for illustrative purposes.

handle outlier (capping)

```python
# Capping
df['Age'] = df['Age'].clip(lower, upper)


import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.boxplot(x=df['Age'])
plt.title('Boxplot of Age after Capping')
plt.show()
```

task 4:

```python
from sklearn.preprocessing import MinMaxScaler


num_cols = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']


scaler = MinMaxScaler()
df_minmax = df.copy()
df_minmax[num_cols] = scaler.fit_transform(df[num_cols])


df_minmax.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_zscore = df.copy()
df_zscore[num_cols] = scaler.fit_transform(df[num_cols])


df_zscore.head()
```

The first table shows the results of Min-Max normalization, where all numerical features are scaled to the range [0, 1].
The second table shows Z-score normalization, where all numerical features have a mean of 0 and a standard deviation of 1.
Both methods ensure that the features are on comparable scales, which is important for machine learning algorithms.
